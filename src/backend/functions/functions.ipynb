```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
```

```python
stocks = ["PG", "JNJ", "PEP", "KO", "MCD", "MSFT", "V"]
```

```python
stocks= open("../../../out/stocks_list.txt").read().split("\n")
```

```python
# Set filtering criteria
MIN_DIVIDEND_YIELD = 2.0  # Minimum dividend yield (%)
MAX_PAYOUT_RATIO = 60.0  # Maximum payout ratio (%)
MIN_DIVIDEND_GROWTH = 5.0  # Minimum 5-year dividend growth rate (%)
```

```python
# Function to fetch stock data
def get_stock_data(ticker):
    try:
        stock = yf.Ticker(ticker)
        return stock.info
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")
        return {}
```

```python
# Retrieve data for all stocks
df = pd.DataFrame([get_stock_data(stock) for stock in tqdm(stocks)])
df
```

```python
def plot_one(df):
    sns.set_theme(style="darkgrid")

    # Normalize 'marketCap' for better bubble sizes
    size_norm = (df["marketCap"] - df["marketCap"].min()) / (
        df["marketCap"].max() - df["marketCap"].min()
    )
    size_scaled = 100 * (size_norm + 0.2)

 

    # Create the scatter plot
    plt.figure(figsize=(12, 8), dpi=500)
    sns.scatterplot(
        data=df,
        x="epsTrailingTwelveMonths",
        y="previousClose",
        hue="industry",
        palette="tab10",
        size=size_scaled,
        sizes=(20, 400),
        edgecolor="black",
        alpha=0.5,
    )

    # Labels and title
    plt.xlabel("EPS (Trailing Twelve Months)", fontsize=12)
    plt.ylabel("Previous Close Price", fontsize=12)
    plt.title("EPS vs Previous Close Price (Colored by Industry)", fontsize=14, fontweight="bold")


    plt.legend(
        title="Industry", 
        bbox_to_anchor=(1.02, 1), 
        loc="upper left", 
        borderaxespad=0.5, 
        frameon=True, 
        fontsize=10,  # Increases sub-legend font size for readability
        title_fontsize=11,  # Makes the title slightly larger
        edgecolor="black",  
        labelspacing=0.8,  # Adds spacing between legend items
        handletextpad=1.2,  # Adjusts spacing between marker and text
        borderpad=0.8,  # Adds padding inside the legend box
    )

    plt.grid(True, linestyle="--", alpha=0.6)

    # Display the plot
    plt.show()

    return None

plot_one(df)
```
